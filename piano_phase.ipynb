```python
from music21 import *

motif_pitches = [
    "E4", "F#4", "B4", "C#5", "D5", "F#4",
    "E4", "C#5", "B4", "F#4", "D5", "C#5"
    ]
melody = stream.Stream()
for p in motif_pitches:
    melody.append(note.Note(p, quarterLength=0.25))
```

```python
melody.show()
```

```python
def create_part(melody, duration_ratio=1, total_cycles=100):
    part = stream.Stream()
    for i in range(total_cycles):
        for n in melody.notes:
            new_note = note.Note(n.nameWithOctave, quarterLength=n.quarterLength * duration_ratio)
            part.append(new_note)
    return part
```

```python
part1 = create_part(melody)
part2 = create_part(melody, duration_ratio=0.99)
```

```python
part1.show()
```

```python
part1.insert(0, instrument.Piano())
part2.insert(0, instrument.Piano())

score = stream.Score()
score.insert(0, tempo.MetronomeMark(number=72))
score.insert(0, part1)
score.insert(0, part2)
score.write("midi", "piano_phase.mid")
```
